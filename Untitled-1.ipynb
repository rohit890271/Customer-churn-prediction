```python
sns.set_theme(style="whitegrid")
quality_counts = df['gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='deeppink')
plt.title('Count Plot of gender')
plt.xlabel('gender')
plt.ylabel('Count')
plt.show()
```
